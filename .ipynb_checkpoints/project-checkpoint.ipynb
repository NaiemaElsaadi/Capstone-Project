## A) Data Cleaning and Preprocessing

<b>A_1) Checking for null values</b>

```python
students.isnull().sum()
```

<B>A_2) Rounding all the float values present inside the dataset into its nearest number</B>

```python
students = students.round()
```

<b>A_3) Converting all the floating values to integer

```python
students[['Admission grade', 'Previous qualification (grade)', 'Curricular units 1st sem (grade)', 'Curricular units 2nd sem (grade)', 'Unemployment rate', 'Inflation rate', 'GDP']] = students[['Admission grade', 'Previous qualification (grade)', 'Curricular units 1st sem (grade)', 'Curricular units 2nd sem (grade)', 'Unemployment rate', 'Inflation rate', 'GDP']].astype(np.int64) 
students.head()
```

<b>A_4) Rename one colums

```python
students = students.rename(columns={'Nacionality': 'Nationality'})
```

```python
students.info()
```

<b>A_5) Dropping all those columns which are not required

```python
students = students.drop(columns=['Application mode', 'Application order', 'International', 'Debtor', 'Marital status', 'Nationality', 'Unemployment rate', 'Inflation rate', 'GDP'], axis=1)
```

```python
students.head()
```

<b>A_6) Changing 'Dropout' to 0, 'Graduate' to 1 and 'Enrolled' to 2 in 'Target' column

```python
students = students.replace({ 'Target' : { 'Dropout' : 0, 'Graduate' : 1, 'Enrolled' : 2 } })
students.head()
```

```python
students.info()
```

```python
#Checking for null values
students.isnull().any()
```

```python
# Handling Missing Values: Replace missing values with the mean of the column
students.fillna(students.mean(), inplace=True)
students.head()
```
